```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display only the first_name column from employees
pd.read_sql("""
SELECT first_name
FROM employees;
""", conn)
```

```python
#Display department_name from department
pd.read_sql("""
SELECT department_name
FROM departments;
""", conn)
```

```python
#Display product_name and price from products
pd.read_sql("""
SELECT product_name, price
FROM products;
""", conn)
```

```python
#Display customer_name and city from customers
pd.read_sql("SELECT * FROM customers LIMIT 5;", conn)
```

```python
#Display employee_id , first name and salary from employees
pd.read_sql("""
SELECT employee_id, first_name, salary
FROM employees;
""", conn)
```
