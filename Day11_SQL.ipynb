```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display all details of the employees whose employee_id is 1
pd.read_sql("""
SELECT *
FROM employees
WHERE employee_id = 1;
""", conn)
```

```python
#Display all products whose category is "Electronics"
pd.read_sql("""
SELECT *
FROM products
WHERE category = 'Electronics';
""", conn)
```

```python
#Display all customers who belong to the city "Bangalore"
pd.read_sql("""
SELECT *
FROM customers
WHERE city = 'Bangalore';
""", conn)
```

```python
#Display all orders placed by customer_id
pd.read_sql("""
SELECT *
FROM orders
WHERE customer_id = 1;
""", conn)
```

```python
#Display first_name, last_name and email of employees who belong to department_id 2
pd.read_sql("""
SELECT first_name, last_name, email
FROM employees
WHERE department_id = 2;
""", conn)
```
