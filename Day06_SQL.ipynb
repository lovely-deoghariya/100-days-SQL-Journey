```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display only the employee_id first_name columns from the first 5 rows of the employees table
pd.read_sql("""
SELECT employee_id, first_name
FROM employees
LIMIT 5;
""", conn)
```

```python
#Display department_id and department_name from the first 4 departments
pd.read_sql("""
SELECT department_id, department_name
FROM departments
LIMIT 4;
""", conn)
```

```python
#Display product_id, product_name and price from the first 6 products
pd.read_sql("""
SELECT product_id, product_name, price
FROM products
LIMIT 6;
""", conn)
```

```python
#Diplay customer_id, customer_name and city from the first 8 customers
pd.read_sql("PRAGMA table_info(customers);", conn)
```

```python
#Display order_id, customer_id, employee_id and order_date from the first 10 orders
pd.read_sql("PRAGMA table_info(orders);", conn)
```

```python
#Display the first 5 rows from the products table showing only product_name and category
pd.read_sql("""
SELECT product_name, category
FROM products
LIMIT 5;
""", conn)
```
