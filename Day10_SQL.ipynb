```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display unique department names from the departments table
pd.read_sql("""
SELECT DISTINCT department_name
FROM departments;
""", conn)
```

```python
#Display unique city values from the customers table
pd.read_sql("""
SELECT DISTINCT city
FROM customers;
""", conn)
```

```python
#Display unique category values from products table
pd.read_sql("""
SELECT DISTINCT category
FROM products;
""", conn)
```

```python
#Display the first 5 unique employee job titles (or another repeating employee attribute if available)
pd.read_sql("""
SELECT DISTINCT job_title
FROM employees
LIMIT 5;
""", conn)
```

```python
#Display the first 10 unique customer cities using DISTINCT AND LIMIT
pd.read_sql("""
SELECT DISTINCT city
FROM customers
LIMIT 10;
""", conn)
```
