```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display employees whose employee_id is BETWEEN 4 AND 8
pd.read_sql("""
SELECT *
FROM employees
WHERE employee_id BETWEEN 4 AND 8;
""", conn)
```

```python
#Display customers whose city is NOT "Bangalore"
pd.read_sql("""
SELECT *
FROM customers
WHERE NOT city = 'Bangalore';
""", conn)
```

```python
#Display products whose category is IN (Electronics Furniture) AND price BETWEEN 1000 and 5000
pd.read_sql("""
SELECT *
FROM products
WHERE category IN ('Electronics', 'Furniture')
AND price BETWEEN 1000 AND 5000;
""", conn)
```

```python
#Display orders where customer_id IN (2,3) OR employee_id = 1
pd.read_sql("""
SELECT *
FROM orders
WHERE customer_id IN (2, 3)
OR order_id = 1;
""", conn)
```

```python
#Display employee_id, first_name and last_name for employees whose department_id IN (1,2) AND employee_id NOT BETWEEN 1 AND 3
pd.read_sql("""
SELECT employee_id, first_name, last_name
FROM employees
WHERE department_id IN (1, 2)
AND employee_id NOT BETWEEN 1 AND 3;
""", conn)
```
