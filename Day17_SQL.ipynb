```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display employees whose employee_id is BETWEEN 1 AND 5
pd.read_sql("""
SELECT *
FROM employees
WHERE employee_id BETWEEN 1 AND 5;
""", conn)
```

```python
#Display customers whose city is IN (Bangalore, Mumbai)
pd.read_sql("""
SELECT *
FROM customers
WHERE city IN ('Bangalore', 'Mumbai');
""", conn)
```

```python
#Display products whose category is NOT Electronics
pd.read_sql("""
SELECT *
FROM products
WHERE NOT category = 'Electronics';
""", conn)
```

```python
#Display orders whose customer_id IN (1,2,3) AND order_id BETWEEN 1
pd.read_sql("""
SELECT *
FROM orders
WHERE customer_id IN (1, 2, 3)
AND order_id BETWEEN 1 AND 20;
""", conn)
```
